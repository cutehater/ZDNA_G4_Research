```python
from OmicsDC import OmicsDataCreate
from pathlib import Path
from OmicsDC import Matching_Experiments_DF
from OmicsDC.resources import ExperimentList
from OmicsDC import Matching_Experiments_Options
import subprocess
import numpy as np
```

```python
experiments = {
    "expid"     :   None,
    "assembly"  :   ["hg38"],
    "Q-value"   :   "50",
    "a.g_class" :   None,
    "antigen"   :   None,
    "cell_type" :   ["Pluripotent stem cell"],
    "cell"      :   None,
    "output"    :   Path("./data_hg_38/tmp")
}

result = OmicsDataCreate(
    expid               = experiments["expid"],
    assembly            = experiments["assembly"],
    assembly_threshold  = experiments["Q-value"],
    antigen_class       = experiments["a.g_class"],
    antigen             = experiments["antigen"],
    cell_type           = experiments["cell_type"],
    cell                = experiments["cell"],
    output_path         = experiments["output"]
    )
```

```python
!tar -xvzf ./data_hg_38/tmp/2025-03-04_16_21_05.tar.gz -C ./data_hg_38/tmp
```

```python
Matching_Experiments_Options = {
                'id'              : None,
                'Genome assembly' : ['hg38'],
                'Antigen class'   : None,
                'Antigen'         : None,
                'Cell type class' : ["Pluripotent stem cell"],
                'Cell type'       : None
                }

Matching_Experiments = Matching_Experiments_DF(ExperimentList, Matching_Experiments_Options)
```

```python
Matching_Experiments = Matching_Experiments[Matching_Experiments['Antigen'].notna()]
```

```python
antigen_class_map = Matching_Experiments.set_index('Antigen')['Antigen class'].to_dict()

for antigen in Matching_Experiments['Antigen'].unique():
    antigen_class = antigen_class_map[antigen]
    antigen_safe = antigen.replace(" ", "_")
    antigen_class_safe = antigen_class.replace(" ", "_")
    
    output_file = f"./data_hg_38/features/stem_cells/{antigen_class_safe}_{antigen_safe}.bed" # replace with your directory name
    command = f"cat ./data_hg_38/tmp/hg38/*_{antigen_safe}_* > {output_file}"
    
    subprocess.Popen(command, shell=True)
```
